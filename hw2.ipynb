```python
diabetes=datasets.load_diabetes()
```

```python
X_age=diabetes.data[:,0] # 나이
X_bmi=diabetes.data[:,2] # BMI
X_p=diabetes.data[:,3] # 혈압
Y = diabetes.target
```

```python
def f(x,y,x1): # regression
    N=len(x)
    X=sum(x)/N
    Y1=sum(y)/N
    X2=sum([i**2 for i in x])/N
    XY=sum(x*y)/N
    a2,b2=(XY-X*Y1)/(X2-X*X),(X2*Y1-X*XY)/(X2-X*X)
    return a2*x1+b2
```

```python
x=np.arange(-1,1,0.1)
plt.figure(figsize=(20,5))

plt.subplot(131)
y=f(X_age,Y,x)
plt.xlim(-0.25,0.25);plt.ylim(0,400);plt.xlabel('x');plt.ylabel('y')
plt.plot(x,y,'k-',label='fitted line');
plt.plot(X_age,Y,'o');
plt.legend();
plt.xlabel('X_age')

plt.subplot(132)
y1=f(X_bmi,Y,x)
plt.xlim(-0.25,0.25);plt.ylim(0,400);plt.xlabel('x');plt.ylabel('y')
plt.plot(x,y1,'k-',label='fitted line');
plt.plot(X_bmi,Y,'o');
plt.xlabel('X_bmi')
plt.legend();

plt.subplot(133)
y2=f(X_p,Y,x)
plt.xlim(-0.25,0.25);plt.ylim(0,400);plt.xlabel('x');plt.ylabel('y')
plt.plot(x,y2,'k-',label='fitted line');
plt.plot(X_p,Y,'o');
plt.legend();
plt.xlabel('X_p')
```

```python
a = LinearRegression() 
a.fit(X_age.reshape(-1,1), Y)
a.fit(X_bmi.reshape(-1,1), Y)
a.fit(X_p.reshape(-1,1), Y)
```

```python
plt.figure(figsize=(20,5))

plt.subplot(131)
plt.plot(X_age, Y, 'o')
plt.plot(X_age,a.predict(X_age.reshape(-1,1)))
plt.xlabel('X_age')

plt.subplot(132)
plt.plot(X_bmi, Y, 'o')
plt.plot(X_bmi,a.predict(X_bmi.reshape(-1,1)))
plt.xlabel('X_bmi')

plt.subplot(133)
plt.plot(X_p, Y, 'o')
plt.plot(X_p,a.predict(X_p.reshape(-1,1)))
plt.xlabel('X_p')
```
